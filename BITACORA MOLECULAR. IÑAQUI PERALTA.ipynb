# Montaje de técnica molecular y tisular en _Bunodosoma sp._


## Resumen

En este trabajo se realizó el montaje de técnicas tisular y molecular, para el procesamiento de la anémona _Bunodosoma sp._ encontrada en Tecolutla, Veracruz; primera vez que se reporta ésta especie para ésta localidad. Se conjuntaron aspectos morfofisiólogicos, histológicos y moleculares, a través de ésto se logró el acercamiento a una identidad taxonómica. Con el diseño de técnicas especie- específicas se llegó a un correcto procesamiento en el laboratorio de esta especie.

## Objetivos


**Objetivo principal**:
* Montaje de técnica tisular y molecular para el correcto procesamiento de la anémona _Bunodosoma sp._ 

**Objetivos secundarios**:
* Reconocer y comparar aspectos morfológicos de las anémonas.
* Caracterizar histológicamente la gastrodermis, disco pedio, disco oral y tentáculos de la anémona para determinar la identidad taxonómica hasta genero de la anémona de mar.
* Montaje de técnica para el procesamiento del gen 16S.

## Actividades realizadas.

**1. Confinamiento de anémona**.

   La búsqueda de información inició con los parámetros fisicoquímicos necesarios para el acuario en los que se determinó los siguientes elementos básicos:

* pH: 7.5 - 8.5.
* Temperatura: 22-26 C.
* Salinidad: 30-34 ‰

Se procedió a la compra de una urna de 70 litros (15 litros por anémona), un espumador, sumidero de limpieza, termómetro, potenciómetro, bomba de agua, calentador, refractómetro, lámpara led para arrecife y la sal marina.

Se prepararon 60 litros de agua con 32 UPS por litro y se dejó madurar por 20 días, revisando diariamente los parámetros antes mencionados.

**2. Colecta y transporte de ejemplares**.

Se visitó la localidad de Tecolutla, Veracruz en octubre del 2017 y en el punto de muestreo 06 “La escollera” (20.4802289 N, -97.0118327000000 O) se recolectaron cuatro ejemplares mediante el siguiente método:

1.	El animal fue identificado en una zona semi-sumergida y con piedras.
2.	Se separó al animal del sustrato utilizando una espátula y creando una ligera corriente con la mano.
3.	El animal fue puesto en una cubeta de 20 litros con agua y arena.

Los ejemplares fueron transportados hacia la Ciudad de México en la cubeta con una bomba de agua y una bomba aireadora de baterías.

**3. Mantenimiento.**

Las anemonas fueron aclimatadas mediante el método de goteo para igualar salinidad, pH y temperatura y tras un periodo de 5 días fueron introducidas a su confinamiento. Las anemonas fueron alimentadas 3 veces por semana con reef snow (Agua purificada, carbonatos de calcio, estroncio, magnesio, potasio, proteínas de origen marino y lípidos.) y en ocasiones con camarón crudo.

Se realizaron cambios de agua parciales del 20% semanalmente para mantener elementos traza al margen.

**4. Reconocimiento morfofisiológico.**

La anemona tiene un cuerpo central de forma cilíndrica en cuya base se encuentra un disco plano que le sirve como pie y para adherirse al sustrato donde luego se desarrollará. En el extremo superior, otro disco denominado disco oral es el que contiene la boca de la anémona de mar, ubicada justo en el centro de éste. Alrededor de la boca se disponen de manera radial una serie de tentáculos en anillos concéntricos. Por lo general cada anillo cuenta con una cantidad de tentáculos en múltiplos de seis.

La boca está conectada al sistema gastrovascular de la anémona por medio de un conducto conocido como actinofaringe. En estas especies cumple la doble función de esófago y faringe. El cuerpo de la anémona de mar es una columna central que puede ser lisa o con algunas protuberancias en las cuales es posible detectar la presencia de nematocistos. La epidermis de estos animales, así como la gastrodermis está formada por una capa gelatinosa conocida como mesoglea que le otorga una enorme plasticidad.

## Resultados

### 1.	Montaje de técnica tisular.

**Anestesia**: Las anémonas fueron anestesiadas con un gradiente descendiente de temperatura hasta alcanzar los 0 °C (Cordero, 2018).

**Toma de muestra**: Se recogieron las muestras a través de una necropsia, separando el organismo en cuatro partes: Tentaculos, disco oral, columna y disco pedio.

**Fijacion**: Se utilizó un fijador oxidante Bouin al 10% por un mes y antes de la inclusión, es conveniente eliminar el ácido pícrico de las muestras (tiñe de color amarillo intenso a los tejidos). El lavado se hace con una solución de alcohol etílico al 70% a la que se añaden algunas gotas de una solución al 5 % de carbonato de litio.

**Inclusion**: Los tejidos fijados adquieren cierta consistencia y dureza, pero no la suficiente para que, de ellos, se obtengan secciones delgadas. Estas secciones, del orden de algunas milésimas de milímetro, se conseguirán cuando los tejidos se infiltren con sustancias denominadas “de inclusión” y adquieran tal dureza que sometidos al filo de una navaja produzcan secciones, cortes o láminas sumamente delgados y transparentes.

**Diagrama de inclusión**:


  

**Microtomia**: Se realizó en un Microtomo de deslizamiento. El movimiento que se realiza para obtener los cortes es horizontal, es decir de atrás hacia adelante y viceversa. Generalmente el mecanismo que sujeta al bloque de parafina es el que se desplaza y a la navaja permanece estática. 

**Extensión y adhesión de los cortes al portaobjetos**: Los cortes obtenidos, aislados o en forma de cintas, generalmente se presentan arrugados y muestran una área menor que la que poseen en la inclusión, por lo que es necesario extenderlos y luego adherirlos a las láminas portaobjetos ( esto facilita su manipulación posterior). Los cortes se extienden al depositarlos sobre la superficie del líquido extendedor contenido en un recipiente denominado “baño de flotación” (éste mantiene la temperatura entre 40 a 45 °C). Las secciones se depositan en el líquido procurando que la superficie brillante (aquella correspondiente al filo de la navaja) se ponga en contacto con el líquido caliente. En caso que ciertas arrugas persistan, se emplean unas pinzas finas o un pincel de pelo de camello, para ejercer una ligera tracción entre los extremos de los cortes y así desarrugarlos totalmente. 

**Tinción**: El método supone la aplicación de la tinción de hematoxilina, que por ser catiónica o básica, tiñe estructuras ácidas (basófilas) en tonos azul y púrpura,como por ejemplo los núcleos celulares; y el uso de eosina que tiñe componentes básicos (acidófilos) en tonos de color rosa, gracias a su naturaleza aniónica o ácida, como el citoplasma.





**Tren de tinción**:


 

**Montaje**: Este procedimiento consiste en colocar encima del corte coloreado y diafanizado una gota de una sustancia adherente, diluida, generalmente en xilol (resina natural como el bálsamo de Canadá o resinas sintéticas, cuyos índices de refracción son similares a los del vidrio) y encima de ellos, una laminilla cubreobjetos, cuidando que no queden burbujas de aire entre la resina. A continuación se deja que el xilol se evapore, y la resina adquiera solidez suficiente; para ello las láminas se colocan en una platina caliente (45 o 50 °C) durante 24 a 48 horas y estarán listas para ser observadas.

### 4.	Montaje de tecnica molecular. 

**Extracción de DNA**: De las muestras se decidió que la mejor región de la anémona para extraer DNA es el disco pedio, ya que es la región que tiene menor presencia de zooxantelas. 

Del kit PROMEGA se realizó la siguiente modificación: 
Agregar 10 μL más de la proteinasa K ya que el tejido no se degradó lo suficiente.

La muestra tardó en degradarse, ya que el tejido estaba muy pegajoso, y tal vez esto interfirió en los resultados para electroforesis y PCR.
